```python
abbr_filepath = os.path.abspath("../data/beck_abbrv_clean.txt")
```

## Read in Abbreviations. Store Abbr in dic

## TODO: check for homographs before storing in dictionary

```python
abbreviation_dic = {}
with codecs.open(abbr_filepath, mode='r', encoding='utf-8') as abbreviations:
    for line in abbreviations.readlines():
        parts = line.split('>')
        parts[0] = parts[0].rstrip(u' ')
        parts[1] = parts[1].lstrip(u' ').rstrip(u'\n')
        tmp_value = []
        if parts[0] in abbreviation_dic.keys():
            tmp_value = abbreviation_dic.get(parts[0])
            if tmp_value is None:
                continue
            abbreviation_dic.update({parts[0]: tmp_value.append(parts[1])})
        else:
            abbreviation_dic.update({parts[0]: [parts[1]]})
```
